Наприклад, давайте напишемо програму, яка визначить, чи є введене число парним чи непарним. Ось код разом з поясненням:

```python
# Запитуємо користувача ввести число
number = int(input("Введіть число: "))

# Перевіряємо, чи є число парним чи непарним
if number % 2 == 0:
    print(f"{number} є парним числом.")
else:
    print(f"{number} є непарним числом.")
```

Пояснення:

Ми використовуємо функцію input(), щоб запитати користувача ввести число. Функція input() приймає введення від користувача у вигляді рядка (строки).

За допомогою int(input()) ми перетворюємо введення користувача у ціле число (integer) і зберігаємо його у змінну number.

Потім ми використовуємо умовний оператор if для перевірки, чи є число парним чи непарним. Умова number % 2 == 0 перевіряє, чи залишок від ділення числа на 2 дорівнює 0. Якщо так, то число є парним, і програма виводить відповідне повідомлення.

Якщо умова number % 2 == 0 не виконується (залишок не дорівнює 0), то число є непарним, і програма виводить відповідне повідомлення.

Ця програма дозволяє користувачеві ввести число і дізнатися, чи воно парне чи непарне. Ви можете вставити цей код в Jupyter Notebook і виконати його, як я показав у попередньому відповіді.
