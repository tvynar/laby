розглянемо послідовні типи даних

list - через те що він mutable його можна змінювати коли захочеш;
tuple - а він є immutable і змінювати його не можна;

```python
l = [1, "2", 3]
ll = list([1, "2", 3])
print("Маємо список, який складається з елементів:","\nl=",l,"\nll=",ll)

l.append(["New", "List"])
print("Тепер список буде:", l)
l.insert(4, "Це буде в кінці")
print("Після вставки елементу на позицію перед 4 будемо мати:", l)

# Видаляємо передостанній елемент та додаємо нові значення
removed_element = l.pop(-2)
removed_element.extend([-1, "тепер це останній елемент"])

print("Елементи масиву зараз:", l)
print("Те що ми видалили з списку та модифікували:", removed_element)
```

Спитались в ChatGPT чому так, ось пояснення:

Цей код має на меті продемонструвати різні операції зі списками в мові програмування Python. Давайте розглянемо його крок за кроком:

1 Створення списків:

l = [1, "2", 3]
ll = list([1, "2", 3])
У цих рядках створюються два однакових списки l і ll, які складаються з трьох елементів: ціле число 1, рядок "2" і ціле число 3.

2 Вивід початкових списків:

print("Маємо список, який складається з елементів:","\nl=",l,"\nll=",ll)
Цей рядок виводить початкові значення обох списків.

3 Додавання елементу вкінці списку:

l.append(["New", "List"])
В цьому рядку до списку l додається новий елемент - список ["New", "List"].

4 Вставка елементу на певну позицію:

l.insert(4, "Це буде в кінці")
Ця операція вставляє рядок "Це буде в кінці" на позицію після четвертого елемента у списку l.

5 Видалення елемента за індексом та модифікація його значення:

removed_element = l.pop(-2)
removed_element.extend([-1, "тепер це останній елемент"])
Цей рядок видаляє передостанній елемент зі списку l за допомогою pop() і зберігає його значення у змінній removed_element. Потім цей видалений елемент, який є списком, розширюється новими значеннями - числом -1 і рядком "тепер це останній елемент".

6 Вивід змінених списків та видаленого елемента:

print("Елементи масиву зараз:", l)
print("Те що ми видалили з списку та модифікували:", removed_element)
Ці рядки виводять змінений список l та вміст змінної removed_element.

Результат виконання цього коду буде таким:

Маємо список, який складається з елементів: 
l= [1, '2', 3, ['New', 'List']]

Тепер список буде: [1, '2', 3, ['New', 'List']]

Елементи масиву зараз: [1, '2', 3, ['New', 'List'], 'Це буде в кінці']
Те що ми видалили з списку та модифікували: ['New', 'List', -1, 'тепер це останній елемент']

знайомство з циклами

for - ітераційний цикл (тобто він повинен пройти задану кількість ітерацій)

while - умовний цикл (ми не знаємо скільки циклів треба виконати, але знаємо умову виходу з циклу)

```python
from random import randint

names = ["Богдан", "Сергій", "Тарас", "Назар", "Василь"]
hobbi = ["футбол", "компютерні ігри", "аніме", "серіали", "хайкінг", "спати"]

for name in names:
    print(f"{name} любить {hobbi[randint(0,5)]}")

print("=|="*10)
# Переписуємо ітераційний цикл за допомогою while
j = 0 #вводимо ітераційну змінну, початкове значення 0 бо ми будемо доступатись до емелентів списку
while j < len(names):
     print(f"{names[j]} любить {hobbi[randint(0,5)]}")
     j+=1

print("=|="*10)

while len(names) > 0:
     print(f"{names.pop()} любить {hobbi[randint(0,5)]}")
```

напишемо програму емуляції кидання кубика (міні казино)

1)буде задано кількість спроб за які нам потрібно викинути бажане значеннея на кубику

2)кидати кубик будемо доки не випаде потрібне число або не закінчиться кількість спроб

3)якщо випало потрібне число в межах кількості спроб то ми виграли

```python
attemps = 3
dice_number = 1 # Значення числа на кубику, від 1 до 6
dice_hit = 0 # початкове значення перед кидання кубика

while dice_hit != dice_number and attemps > 0:
    dice_hit = randint(1,6) # Це буде симулювати кидання кубика, нам буде випадати випадкове число від 1 до 6
    attemps -= 1
    print(f"Випало значення {dice_hit}, воно відповідає бажаному: {dice_hit == dice_number}.")
    if dice_hit == dice_number:
        print("Ви виграли!")
    elif attemps == 0:
        print("Ви програли!")
    else:
        print(f"Спробуйте ще, залишилось {attemps} спроб.")
```

```python
def check_if_value_is_none(value:str):
    _ = value == None
    print(f"Перевіряємо чи значення {value} є рівним None: {_}")
    return _
    print("Це ніколи не виконається бо ми вже вийши з функції!")

check_if_value_is_none("Yes")
check_if_value_is_none("None")
check_if_value_is_none(None)
```

```python
# Функції можна викликати або передавати як об'єкти

f= check_if_value_is_none
w= check_if_value_is_none ("Неважливо")
print(f"Це є НЕ викликана функція яка записана у змінну:{f}\nA це вже результат який повертає виклакану функцію: {w}")
print(f("передали у функцію!"))


def pass_fun_inside_fun(fun):
    print("Це код основної функції")
    fun()
    print("Це вже буде після виклику переданої функції")

def inside_fun():
    print("Це вивід з функції яку ми передаємо в іншу фукнцію")

pass_fun_inside_fun (inside_fun)
```

Спробуємо написати Лямбду

```python
# прототип Лямбди
def lambda_prototype(p:int):
    return [p + i if p > 10 else p - i for i in range(p)]

print(f"Це виклик прототипу Лямбди: {lambda_prototype(11)}")

# а тепер перепишемо це через Лямбду
lm = lambda p: [p + i if p > 10 else p - i for i in range(p)]

print(f"А це вже виклик Лямбди: {lm(9)}")
```
